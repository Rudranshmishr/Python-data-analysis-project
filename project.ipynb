# Exploratory Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x='Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=True)
```

```python
sales=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=True)
ax=sns.barplot(x='Gender',y='Amount',data=sales)

for bars in ax.containers:
    ax.bar_label(bars)
```

# From the above graph we can conclude that purchasing power of female is greater then the male 

# age

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
ax=sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

# From the above graph we can conclude that most of the buyers are from the age group of 26-35

# state

```python
sales_st=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Orders',data=sales_st)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_st=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Amount',data=sales_st)


```

# From above graph we can conclude that Uttar Pradesh, Maharashtra , Karnataka are the top 3 state from the sales point of view

# Marital_Status

```python
ax=sns.countplot(x='Marital_Status',data=df)

for bars in ax.containers:
    ax.bar_label(bars)



```

```python
sales_ms=df.groupby(['Marital_Status','Gender' ], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_ms)
```

from the above graph we can conclude that most the buyers are married (female) and the have high purchasing power

# occupation


```python
sns.set({'figure.figsize':(20,6)})
ax=sns.countplot(x='Occupation', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
ax=sns.barplot(x='Occupation',y='Amount',data=sales_occ)
```

From the above graph we can conclude that most of buyers are working in IT sector,Healthcare,Aviation,Banking sector

# Product Category

```python
sns.set({'figure.figsize':(25,10)})
ax=sns.countplot(x='Product_Category',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pc=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
ax=sns.barplot(x='Product_Category',y='Amount',data=sales_pc)
```

From above graph we can conclude that most of the sold product are from food,clothing and foorwear
